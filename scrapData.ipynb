```python
import requests
from bs4 import BeautifulSoup
import csv

def scrape_flipkart_pages(base_url):
    all_products = []

    page_number = 1
    while True:
        search_url = base_url + f'&page={page_number}'
        response = requests.get(search_url)
        soup = BeautifulSoup(response.content, 'html.parser')

        products_data = soup.find_all('div', class_='_2WkVRV')
        tags_data = soup.find_all('a', class_='IRpwTa')

        if not products_data or not tags_data:
            break

        for product, tag in zip(products_data, tags_data):
            product_name = product.text
            product_tag = tag.text
            product_link = 'https://www.flipkart.com' + tag['href']

            all_products.append({'name': product_name, 'tag': product_tag, 'link': product_link})

        page_number += 1

    return all_products

def save_to_csv(data):
    with open('product_name_and_tags.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=['name', 'tag', 'link'])
        writer.writeheader()
        writer.writerows(data)

if __name__ == '__main__':
    base_url = 'https://www.flipkart.com/search?q=clothing&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
    product_data = scrape_flipkart_pages(base_url)
    save_to_csv(product_data)
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import time

def scrape_flipkart_pages(base_url):
    all_products = []

    with requests.Session() as session:
        page_number = 1
        while page_number <= 892:  # Total number of pages
            search_url = base_url + f'&page={page_number}'
            try:
                response = session.get(search_url)
                response.raise_for_status()  # Raise exception for HTTP error responses
                soup = BeautifulSoup(response.content, 'html.parser')

                products_data = soup.find_all('div', class_='_2WkVRV')
                tags_data = soup.find_all('a', class_='IRpwTa')

                if not products_data or not tags_data:
                    break

                for product_data, tag_data in zip(products_data, tags_data):
                    product_name = product_data.text
                    product_tag = tag_data.text
                    product_link = 'https://www.flipkart.com' + tag_data['href']

                    all_products.append({'name': product_name, 'tag': product_tag, 'link': product_link})

                print(f"Scraped {len(all_products)} products from page {page_number}")
                page_number += 1
                time.sleep(2)  # Adding a small delay to avoid overloading the server

            except requests.exceptions.RequestException as e:
                print(f"Error: {e}")
                print("Retrying after 5 seconds...")
                time.sleep(5)

    return all_products

def save_to_csv(data):
    with open('all_products.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=['name', 'tag', 'link'])
        writer.writeheader()
        writer.writerows(data)

if __name__ == '__main__':
    base_url = 'https://www.flipkart.com/clothing-and-accessories/pr?sid=clo&marketplace=FLIPKART&otracker=product_breadCrumbs_Clothing+and+Accessories'
    product_data = scrape_flipkart_pages(base_url)
    save_to_csv(product_data)
```
